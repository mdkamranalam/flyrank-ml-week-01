## 2. Target or proxy

We are predicting whether a page is losing search performance.
Using the starter dataset, our proxy target is an observed outcome: whether the page is currently in a downward trend.
We define this as `is_declining_label = (trend_direction == 'down')`.
(Note: The target must be an observed outcome, not a product-defined rule like `health_score`).

```python
# Check the target distribution in starter data
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv')
is_declining = (df['trend_direction'] == 'down')
print("Target proxy distribution (is_declining_label):")
print(is_declining.value_counts(normalize=True).round(3))
```
